```python
import torch
import numpy as np
import pandas as pd
data_np = pd.read_csv('weatherAUS.csv')
data_tensor = torch.from_numpy(data_np.iloc[:100, 2:7].values.astype(np.float32))

print(f"PyTorch Tensor形状: {data_tensor.shape}")
print(f"数据类型: {data_tensor.dtype}")

# 2. 基础操作（今天掌握这3个即可）
print(f"平均值: {data_tensor.mean()}")
print(f"标准差: {data_tensor.std()}")

# 3. 自动求导（Autograd，PyTorch核心特性）
x = torch.tensor([2.0, 3.0], requires_grad=True)
y = x.pow(2).sum()  # y = x1² + x2²
y.backward()
print(f"梯度: {x.grad}")  # 应该是 [4.0, 6.0]（2x）
```
