```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('weatherAUS.csv')
print(data.head(3))
temps = data['Temp3pm'].dropna().values
print(f"\n温度数据:{temps[:10]}")
print(f"\n平均温度:{np.mean(temps).round(2)}")
plt.figure(figsize=(10,6))
plt.hist(temps,bins=30,color='blue',edgecolor='black')
plt.title('afternoon temperature distribution(3pm)',fontsize=20)
plt.xlabel('temperature',fontsize=12)
plt.ylabel('frequency',fontsize=12)
plt.grid(axis='y',alpha=0.75)
plt.savefig('temp_distribution.png',dpi=300,bbox_inches='tight')
plt.show()
print(f"\n样本统计:")
print(f"样本数:{len(temps)}")
print(f"平均值:{np.mean(temps).round(2)}")
print(f"标注差:{np.std(temps).round(2)}")
print(f"最高温:{np.max(temps).round(2)}")
print(f"最低温:{np.min(temps).round(2)}")
```
